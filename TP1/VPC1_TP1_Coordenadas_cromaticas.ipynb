##### ALGORITMO DE COORDENADAS CROMATICAS

```python
#Si queremos que las imágenes sean mostradas en una ventana emergente quitar el inline
%matplotlib inline
# OpenCV-Python utiliza NumPy para el manejo de imágenes
import numpy as np
# cv2 es el módulo python para acceder a OpenCV 
import cv2 as cv
# Usamos las poderosas herramientas de graficación de matplotlib para mostrar imágenes, perfiles, histogramas, etc
import matplotlib.pyplot as plt
```

```python

# Funcion para pasar de rgb a coordenadas cromaticas
def rgb2cromatica(imagen):
    imagen = imagen.astype(np.uint16)
    imagen_b, imagen_g, imagen_r = cv.split(imagen)
    for i in range(imagen.shape[0]):
        for j in range(imagen.shape[1]):
            div = (imagen_b[i,j]+imagen_g[i,j]+imagen_r[i,j])
            if div > 0:
                imagen_b[i,j] =  (imagen_b[i,j] / div)*255
                imagen_g[i,j] =  (imagen_g[i,j] / div)*255
                imagen_r[i,j] =  (imagen_r[i,j] / div)*255
    new_imagen = cv.merge((imagen_b,imagen_g,imagen_r))
    new_imagen = new_imagen.astype(np.uint8)
    return new_imagen
```

```python
# Cargamos las imagenes a corvertir de RGB a coordenadas cromaticas
img1 = cv.imread('C:/Users\juana\OneDrive\Documentos\IA\CEIA\VC1\Material_TPs\TP1\coord_cromaticas\CoordCrom_1.png')
img2 = cv.imread('C:/Users\juana\OneDrive\Documentos\IA\CEIA\VC1\Material_TPs\TP1\coord_cromaticas\CoordCrom_2.png')
img3 = cv.imread('C:/Users\juana\OneDrive\Documentos\IA\CEIA\VC1\Material_TPs\TP1\coord_cromaticas\CoordCrom_3.png')
```

```python
# Pasamos de rgb a coordenadas cromaticas cada una de las imagenes
new_img1 = rgb2cromatica(img1)
new_img2 = rgb2cromatica(img2)
new_img3 = rgb2cromatica(img3)
```

```python
# Ploteamos las figuras en coordenadas cromaticas
plt.figure(1,figsize=(20,15))
plt.subplot(1,3,1)
plt.imshow(cv.cvtColor(new_img1, cv.COLOR_BGR2RGB))
plt.subplot(1,3,2)
plt.imshow(cv.cvtColor(new_img2, cv.COLOR_BGR2RGB))
plt.subplot(1,3,3)
plt.imshow(cv.cvtColor(new_img3, cv.COLOR_BGR2RGB))
plt.show()
```

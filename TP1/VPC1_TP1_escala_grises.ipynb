```python
#Si queremos que las imágenes sean mostradas en una ventana emergente quitar el inline
%matplotlib inline
# OpenCV-Python utiliza NumPy para el manejo de imágenes
import numpy as np
# cv2 es el módulo python para acceder a OpenCV 
import cv2 as cv
# Usamos las poderosas herramientas de graficación de matplotlib para mostrar imágenes, perfiles, histogramas, etc
import matplotlib.pyplot as plt
```

```python
# Cargamos las imagenes en escala de grises 
img1_tp = cv.imread('C:/Users\juana\OneDrive\Documentos\IA\CEIA\VC1\Material_TPs\TP1/img1_tp.png',0)
img2_tp = cv.imread('C:/Users\juana\OneDrive\Documentos\IA\CEIA\VC1\Material_TPs\TP1/img2_tp.png',0)
```

```python

# Ploeteamos las imagenes
plt.figure(1,figsize=(20,8))
plt.subplot(1,2,1)
plt.title('Imagen 1')
plt.imshow(img1_tp,cmap='gray')
plt.subplot(1,2,2)
plt.title('Imagen 2')
plt.imshow(img2_tp,cmap='gray')
plt.show()
```

```python
# Graficamos los histogramas de cada una de las imagenes
hist1,bins1 = np.histogram(img1_tp.ravel(),256,[0,256])
hist2,bins2 = np.histogram(img2_tp.ravel(),256,[0,256])
plt.figure(1,figsize=(20,8))
plt.subplot(1,2,1)
plt.title('Histograma imagen 1')
plt.plot(hist1)
plt.subplot(1,2,2)
plt.title('Histograma imagen 2')
plt.plot(hist2)
plt.show()
```

Se observa que ambos histogramas son exactamente iguales, lo que quiere decir que las imagenes tinene los mismos niveles de intensidad de pixeles, solo que estan cada uno de ellos ubicados en distinta posicion dentro de la imagen. 

Esto quiere decir que no podemos tomar como "feature" a los histogramas de imagenes en escala de grises para entrenar modelos de prediccion/clasificacion, ya que un mismo histograma corresponde a 2 imagenes completamente distintas.

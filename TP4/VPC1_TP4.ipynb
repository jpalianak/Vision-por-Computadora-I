#### FUNCION DE LBP

```python
def LBP_image(img):
    LBP = np.zeros([img.shape[0]-1, img.shape[1]-1], dtype=int)
    for i in np.arange(1, img.shape[0]-1):
        for j in np.arange(1, img.shape[1]-1):
            inten = img[i,j]
            
            if img[i-1,j] >= inten:
                LBP[i,j] += 2**0

            if img[i-1,j+1] >= inten:
                LBP[i,j] += 2**1

            if img[i,j+1] >= inten:
                LBP[i,j] += 2**2

            if img[i+1,j+1] >= inten:
                LBP[i,j] += 2**3

            if img[i+1,j] >= inten:
                LBP[i,j] += 2**4

            if img[i+1,j-1] >= inten:
                LBP[i,j] += 2**5

            if img[i,j-1] >= inten:
                LBP[i,j] += 2**6

            if img[i-1,j-1] >= inten:
                LBP[i,j] += 2**7
  

    return LBP
```

```python
LBP_img1 = LBP_image(img1_gray)
LBP_img2 = LBP_image(img2_gray)
LBP_img3 = LBP_image(img3_gray)
LBP_img4 = LBP_image(img4_gray)
LBP_img5 = LBP_image(img5_gray)

hist1,bins1 = np.histogram(LBP_img1.ravel(),256,[0,256])
hist2,bins2 = np.histogram(LBP_img2.ravel(),256,[0,256])
hist3,bins3 = np.histogram(LBP_img3.ravel(),256,[0,256])
hist4,bins4 = np.histogram(LBP_img4.ravel(),256,[0,256])
hist5,bins5 = np.histogram(LBP_img5.ravel(),256,[0,256])

plt.figure(figsize=(25,10))
plt.subplot(3,5,1),plt.imshow(img1_rgb),plt.subplot(3,5,6),plt.imshow(LBP_img1,cmap='gray'),plt.subplot(3,5,11),plt.plot(hist1)
plt.subplot(3,5,2),plt.imshow(img2_rgb),plt.subplot(3,5,7),plt.imshow(LBP_img2,cmap='gray'),plt.subplot(3,5,12),plt.plot(hist2)
plt.subplot(3,5,3),plt.imshow(img3_rgb),plt.subplot(3,5,8),plt.imshow(LBP_img3,cmap='gray'),plt.subplot(3,5,13),plt.plot(hist3)
plt.subplot(3,5,4),plt.imshow(img4_rgb),plt.subplot(3,5,9),plt.imshow(LBP_img4,cmap='gray'),plt.subplot(3,5,14),plt.plot(hist4)
plt.subplot(3,5,5),plt.imshow(img5_rgb),plt.subplot(3,5,10),plt.imshow(LBP_img5,cmap='gray'),plt.subplot(3,5,15),plt.plot(hist5)
```
